```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

# Function

```python
# Define a linear function
def linear_function(x): 
    return 0.5*x**2 + 1*x + 0
# numpy list of length 10
function_input_list = np.linspace(0, 10, 10)
# output y produced by linear function 
function_output_list = linear_function(function_input_list) 
# Will plot the inputs and outputs on the graph
plt.figure(1, dpi=120) 
plt.plot(function_input_list, function_output_list, "o", label="Function")
plt.legend()
```

# Visualizing Numerical Derivatives

```python
# Derivative function
def derivative(function_input_list, function_output_list):
    dy = np.diff(function_output_list)
    dx = np.diff(function_input_list)
    dydx  = dy/dx
    x = []
    for i in range(len(dydx)):
        xtemp = ( function_input_list[i] + function_input_list[i+1] )/2
        x = np.append(x, xtemp)
    return x, dydx
# output of derivate
derivative_input_list, derivative_output_list = derivative(function_input_list, function_output_list)
# graph of derivative
plt.figure(1, dpi=120) 
plt.plot(derivative_input_list, derivative_output_list, "o", label="Derivative")
plt.legend()
```

# Visualizing Numerical Integration

```python
from scipy.integrate import quad
# Derivative function
def integration(function_input_list):    
    x = []
    y = []
    for i in range(len(function_input_list)-1):
        # call quad to integrate f from 0 to i+1
        res, err = quad(linear_function, 0, i+1)
        xtemp = ( function_input_list[i] + function_input_list[i+1] )/2
        x = np.append(x, xtemp)
        y = np.append(y, res)
    return x, y
# output of integration
integration_input_list, integration_output_list = integration(function_input_list)
# graph of integration
plt.figure(1, dpi=120) 
plt.plot(integration_input_list, integration_output_list, "o", label="Integration")
plt.legend()
```

# Visualizing all in one graph

```python
plt.figure(1, dpi=120) 
plt.plot(function_input_list, function_output_list, "o", label="Function")
plt.plot(derivative_input_list, derivative_output_list, "o", label="Derivative")
plt.plot(integration_input_list, integration_output_list, "o", label="Integration")
plt.legend()
```
